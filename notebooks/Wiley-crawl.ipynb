```python
import glob
import json



data_dir = '../data/wiley-crawl/'

files = glob.glob(data_dir + '*.html')
files
```

```python
from scripts.wiley_crawl import read_file_soup, extract_issue_records

records = []
for fname in files:
    issue_soup = read_file_soup(fname)
    try:
        records.extend(extract_issue_records(issue_soup))
    except ValueError:
        print(fname)
        raise


len(records)
```

```python
records_file = '../data/wiley-crawl-records.json'
with open(records_file, 'wt') as fh:
    json.dump(records, fh)
```
