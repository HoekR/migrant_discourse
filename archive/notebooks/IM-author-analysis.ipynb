```python
# the name and location of the article records for the IM journal (in CSV format)
records_file = '../data/wiley-IM-records.json'

# load the csv data into a data frame
df = pd.read_json(records_file)
# show the first and last records of the dataset to demonstrate what the records look like
df
```

```python
for row in df[df.article_author.isna()][['article_doi', 'article_title', 'issue_pub_year']].iterrows():
    doi, title, year = row[1]
    if year != 1975:
        continue
    print(row[1][0], row[1][1], row[1][2])
```

```python
show_columns = ['issue_pub_year', 'article_doi', 'article_title']
for row in df[(df.issue_section == 'Experts/Expertos')][show_columns].iterrows():
    year, doi, title = row[1]
    if year < 1961 or year > 1999:
        continue
    print(row[1][0], row[1][1], row[1][2])
```

```python
counts = df[df.issue_pub_year < 1984].issue_section.value_counts()
counts[counts > 5]
```

```python
from collections import Counter


author_string_list = list(df[df.article_author.isna() == False].article_author)

author_count = Counter()

for author_string in author_string_list:
    authors = author_string.split(' & ')
    author_count.update([author.title() for author in authors])
    
for author, count in author_count.most_common():
    print(f'{author: <30}{count: >3}')
```
